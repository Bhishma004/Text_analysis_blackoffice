```python
import os
import pandas as pd
from myfunction import txt_analysis
```

```python
# Define the file paths
excel_file_path = "Output Data Structure.xlsx"
my_file_folder = "txt_file"

# Read the Excel file into a DataFrame
df = pd.read_excel(excel_file_path)

# Iterate through files in the my_file folder
for file_name in os.listdir(my_file_folder):
    if file_name.endswith(".txt"):
        file_path = os.path.join(my_file_folder, file_name)

        # Initialize txt as an empty string
        txt = ""

        # Open the file and read its content
        try:
            with open(file_path, "r", encoding="utf-8") as file:
                txt = file.read()
        except FileNotFoundError:
            print(f"File '{file_path}' not found.")
            continue

        # Add a space after periods
        txt = txt.replace('.', '. ')

        # Analyze the text
        t = txt_analysis(txt)

        # Extract the filename from the txt_file_path (e.g., "321.txt" -> "321")
        filename, _ = os.path.splitext(file_name)

        # Update the corresponding row in the DataFrame with the scores
        try:
            # Convert filename to float
            url_id = float(filename)
            mask = df['URL_ID'] == url_id
            if mask.any():
                index = df.index[mask][0]
                df.at[index, 'POSITIVE SCORE'] = t[0]
                df.at[index, 'NEGATIVE SCORE'] = t[1]
                df.at[index, 'POLARITY SCORE'] = t[2]
                df.at[index, 'SUBJECTIVITY SCORE'] = t[3]
                df.at[index, 'AVG SENTENCE LENGTH'] = t[4]
                df.at[index, 'PERCENTAGE OF COMPLEX WORDS'] = t[5]
                df.at[index, 'FOG INDEX'] = t[6]
                df.at[index, 'AVG NUMBER OF WORDS PER SENTENCE'] = t[7]
                df.at[index, 'COMPLEX WORD COUNT'] = t[8]
                df.at[index, 'WORD COUNT'] = t[9]
                df.at[index, 'SYLLABLE PER WORD'] = t[10]
                df.at[index, 'PERSONAL PRONOUNS'] = t[11]
                df.at[index, 'AVG WORD LENGTH'] = t[12]
            else:
                print(f"URL_ID {url_id} not found in the Excel file.")
        except ValueError:
            print(f"Invalid URL_ID: {filename}")

# Save the updated DataFrame, including the 'URL_ID' and 'URL_LINK' columns
df.to_excel(excel_file_path, index=False)

print("Positive and Negative Scores updated in the Excel file, 'URL_ID' and 'URL_LINK' columns preserved.")
```
